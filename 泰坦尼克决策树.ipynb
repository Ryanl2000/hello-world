```python
import numpy as np
import pandas as pd
from sklearn import tree
from sklearn.model_selection import *
import matplotlib.pyplot as plt
from sklearn.datasets import *
```

```python
path_test = r"C:\Users\Luogl2000\Desktop\数分\titanic\test.csv"
path_train = r"C:\Users\Luogl2000\Desktop\数分\titanic\train.csv"
```

```python
train_data = pd.read_csv(path_train)
```

```python
train_data.info()
```

```python
train_data.drop(['Cabin','Name','Ticket'],axis = 1,inplace = True)
```

```python
train_data
```

```python
train_data['Age'].fillna(train_data['Age'].mean(),inplace = True)
```

```python
train_data = train_data.dropna()
```

```python
Emb_lst = list(train_data.Embarked.unique())
```

```python
Emb_lst
```

```python
train_data.Embarked = train_data.Embarked.apply(lambda x :Emb_lst.index(x))
```

```python
train_data['Sex'] = (train_data['Sex'] == 'male').astype(int)
```

```python
train_data
```

```python
X_train = train_data.iloc[:,train_data.columns != 'Survived']
Y_train = train_data.iloc[:,train_data.columns == 'Survived']
```

```python
test_data = pd.read_csv(path_test)
```

```python
test_data
```

```python
test_data.drop(['Cabin','Name','Ticket'],axis = 1,inplace = True)
```

```python
test_data.info()
```

```python
test_data['Age'].fillna(test_data['Age'].mean(),inplace = True)
```

```python
test_data = test_data.dropna()
```

```python
test_data['Embarked'] = test_data['Embarked'].apply(lambda x :Emb_lst.index(x))
test_data['Sex'] = (test_data['Sex'] == 'male').astype(int)
```

```python
test_data
```

```python
X_test = test_data
```

```python
clf = tree.DecisionTreeClassifier(criterion = 'entropy'
                                 ,random_state = 25
                                 ,splitter = 'best'
                                 ,max_depth = 3
                                 )
clf = clf.fit(X_train,Y_train)
```

```python
Y_test = pd.DataFrame(clf.predict(X_test))
```

```python
Y_test
```

```python
Y_test.to_csv(r"C:\Users\Luogl2000\Desktop\数分\titanic\exp.csv")
```
